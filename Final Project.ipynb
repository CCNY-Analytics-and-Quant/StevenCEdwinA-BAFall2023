```python
# Credit to Steven Fernandez's code.

# Convert 'order_date_DateOrders' column to datetime
dfOD2['order_date_DateOrders'] = pd.to_datetime(dfOD2['order_date_DateOrders'])

# Select columns and convert 'Order_Profit_Per_Order' to numeric using .loc
profitOverTime = dfOD2[['order_date_DateOrders', 'Order_Profit_Per_Order']].copy()
profitOverTime.loc[:, 'Order_Profit_Per_Order'] = pd.to_numeric(profitOverTime['Order_Profit_Per_Order'])

# Set 'order_date_DateOrders' as the index
profitOverTime.set_index('order_date_DateOrders', inplace=True)

# Ensure the index is a DatetimeIndex
profitOverTime.index = pd.to_datetime(profitOverTime.index)

# Resample by month ('M') and sum the profits
profitTrend = profitOverTime['Order_Profit_Per_Order'].resample('M').sum()

ProfitTrend = profitOverTime['Order_Profit_Per_Order'].resample('M').sum()
ProfitTrend.head
```

```python
# Visualization on how the profit has changed from 2015 - 2018. Can see that profits 9/2017 and then started to decline rapidly.
plt.figure(figsize=(24, 12))
plt.plot(profitTrend.index, profitTrend.values, label='Monthly Profit Trend Line')
plt.title('Profit Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)
plt.show()
```
