```python
# Summing up product prices and sorted by regions from greatest amount of money made to least. Western Europe has the greatest value of total sales while Western Europe has the least.
# Group by 'Region' and sum the 'Sales' for each region
region_sales_sum = dfOD.groupby('Order_Region')['Sales'].sum().reset_index()

# Sort the summed sales values in descending order
sorted_sales = region_sales_sum.sort_values(by='Sales', ascending=False)

# Identify regions that made the most and least money
region_most_money = sorted_sales.iloc[0]  # Region with the most money
region_least_money = sorted_sales.iloc[-1]  # Region with the least money

print("Region making the most money:")
print(region_most_money)
print("\nRegion making the least money:")
print(region_least_money)
```

```python
# Visualization of the distribution of total sales per region.
plt.figure(figsize=(10, 6))
plt.bar(sorted_sales['Order_Region'], sorted_sales['Sales'])
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.xlabel('Regions')
plt.title('Total Sales by Region')
plt.tight_layout()
plt.show()
```
