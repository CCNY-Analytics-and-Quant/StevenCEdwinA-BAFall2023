```python
# Loading dataframe again so I don't have to scroll up
dfOD = pd.read_csv('Results_OrderData.csv')
dfOD
```

```python
# Summing up product prices and sorted by regions from greatest amount of money made to least. Western Europe has the greatest value of total sales while Western Europe has the least.
# Group by 'Region' and sum the 'Sales' for each region
region_sales_sum = dfOD.groupby('Order_Region')['Sales'].sum().reset_index()

# Sort the summed sales values in descending order
sorted_sales = region_sales_sum.sort_values(by='Sales', ascending=False)

# Identify regions that made the most and least money
region_most_money = sorted_sales.iloc[0]  # Region with the most money
region_least_money = sorted_sales.iloc[-1]  # Region with the least money

print("Region making the most money:")
print(region_most_money)
print("\nRegion making the least money:")
print(region_least_money)
```

```python
# Visualization of the distribution of total sales per region.
plt.figure(figsize=(10, 6))
plt.bar(sorted_sales['Order_Region'], sorted_sales['Sales'])
plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.xlabel('Regions')
plt.title('Total Sales by Region')
plt.tight_layout()
plt.show()
```

```python
# Realization that the 'Order_Proft_Per_Order' category has some negatives, will need to create visualization to the trend.
dfOD2 = pd.read_csv('Results_OrderData2.csv')
dfOD2 
```

```python
# Credit to Steven Fernandez's code.

# Convert 'order_date_DateOrders' column to datetime
dfOD2['order_date_DateOrders'] = pd.to_datetime(dfOD2['order_date_DateOrders'])

# Select columns and convert 'Order_Profit_Per_Order' to numeric using .loc
profitOverTime = dfOD2[['order_date_DateOrders', 'Order_Profit_Per_Order']].copy()
profitOverTime.loc[:, 'Order_Profit_Per_Order'] = pd.to_numeric(profitOverTime['Order_Profit_Per_Order'])

# Set 'order_date_DateOrders' as the index
profitOverTime.set_index('order_date_DateOrders', inplace=True)

# Ensure the index is a DatetimeIndex
profitOverTime.index = pd.to_datetime(profitOverTime.index)

# Resample by month ('M') and sum the profits
profitTrend = profitOverTime['Order_Profit_Per_Order'].resample('M').sum()

profitTrend = profitOverTime['Order_Profit_Per_Order'].resample('M').sum()
profitTrend.head
```

```python
# Visualization on how the profit has changed from 2015 - 2018. Can see that profits 9/2017 and then started to decline rapidly.
plt.figure(figsize=(24, 12))
plt.plot(profitTrend.index, profitTrend.values, label='Monthly Profit Trend Line')
plt.title('Profit Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Loading the uncleaned data back in for cleaning due to cleaning on pandas being more reliable than SQL. 
dfODA = pd.read_csv('Results_OrderDataAll.csv')
dfODA
```

```python
# Cleaning what's unnecessary to figure out the next step. Which is the item distribution for most ordered items.
columnstoDrop = ['column1', 'Benefit_per_order', 'Order_City', 'Order_Country', 'Order_Customer_Id', 'Order_Id', 'Order_Item_Cardprod_Id', 'Order_Item_Profit_Ratio', 'Order_State', 'Order_Zipcode']
dfODA.drop(columnstoDrop, axis = 1, inplace = True)
dfODA
```

```python
# Checking what the most sold products are

# Reloading Product Data to reduce the amount of scrolling required. Product Data will tell us the product name and category it belongs in.
dfPD = pd.read_csv('Results_ProductData.csv')
dfPD
```

```python
# Checking what the most sold products are
itemCounts = dfPD['Category_Name'].value_counts()

# Calculate percentages | Electronics makes up ~9.3% of the distribution
percentages = (itemCounts / itemCounts.sum()) * 100

# Sort the data in descending order
percentages = percentages.sort_values(ascending = True)

# Plotting the horizontal bar graph
fig, ax = plt.subplots(figsize = (20, 12))
bars = ax.barh(percentages.index, percentages)

# Separating the data from the top 5 most produced items compared to the middle and bottom.
top_5 = percentages.nlargest(5).index.to_list()
bot_5 = percentages.nsmallest(26).index.to_list()

for i, bar in enumerate(bars):
    if percentages.index[i] in top_5:
        bar.set_color('green')
    elif percentages.index[i] in bot_5:
        bar.set_color('red')

# Plotting a horizontal bar graph
plt.xlabel('Percentage')
plt.title('Distribution of Items')
plt.show()

print('It can be seen that the top 5 most sold products are electronics, kids golf clubs, golf gloves, accessories, and golf balls.')
```

```python
# Now need to figure out the actual costs from the data. But there's no explicit data for it so we have to do some math.
productCost = dfODA[['order_date_DateOrders', 'Order_Region', 'Order_Profit_Per_Order', 'Order_Item_Total']]
productCost
```

```python
# Item Total: Revenue - Profit = Cost (Warning is fine)
productCost['Cost'] = productCost['Order_Item_Total'] - productCost['Order_Profit_Per_Order']
productCost
```

```python
# Ran into a consistent issue where KeyError: 'order_date_DateOrders' kept occurring. Used to check the index of the dataframe productCost. Some errors can be solved through running the right cells (KEEP IN MIND)
print(productCost.columns)
```

```python
# Graph code to see a visualization on the relationships between when the sales and costs started dropping dramatically.
productCost['order_date_DateOrders'] = pd.to_datetime(productCost['order_date_DateOrders'])

# Set 'order_date' as the index
productCost.set_index('order_date_DateOrders', inplace=True)

# Resample data to monthly frequency and calculate the sum
monthly_data = productCost.resample('M').sum()

# Plotting
plt.figure(figsize=(12, 6))

# Plot monthly sales
plt.plot(monthly_data.index, monthly_data['Order_Item_Total'], label='Monthly Sales', marker='o')

# Plot monthly costs
plt.plot(monthly_data.index, monthly_data['Cost'], label='Monthly Costs', marker='o')

plt.title('Monthly Sales & Costs of Orders Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

```python
# Filtering shipping data to see which market, status, and shipping mode deliveries are
df_shipping_data_all = pd.read_csv('Results_ShippingDataAll2.csv')
deliveryStatus = df_shipping_data_all[['Delivery_Status', 'Shipping_Mode', 'Market']]
deliveryStatus
```

```python
# Filter data for 'Shipping on time' deliveries
on_time_deliveries = deliveryStatus[df_shipping_data_all['Delivery_Status'] == 'Shipping on time']

# Group by 'Market' and count on-time deliveries
on_time_counts = on_time_deliveries.groupby('Market')['Delivery_Status'].count().reset_index(name='On Time Deliveries')

# Print to check the content of on_time_counts DataFrame
print(on_time_counts)
print('\n')

# Sort markets by the count of on-time deliveries in descending order
on_time_counts = on_time_counts.sort_values(by='On Time Deliveries', ascending=False)

# Print to check the sorted DataFrame
print(on_time_counts)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(on_time_counts['Market'], on_time_counts['On Time Deliveries'], color='green')
plt.ylabel('Count of On-Time Deliveries')
plt.title('Count of On-Time Deliveries by Market')
plt.xticks(rotation=360, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# The Results_ShippingDataAll.csv file kept giving me an empty dataset. Had to go back into SQL and get a fresh csv file.

# Filter data for 'Late delivery' in column 'Delivery_Status' in dataframe 'df_shipping_data_all'.
late_deliveries = deliveryStatus[df_shipping_data_all['Delivery_Status'] == 'Late delivery']

# Group by 'Market' and late deliveries
late_delivery_counts = late_deliveries.groupby('Market')['Delivery_Status'].count().reset_index(name='Late Deliveries')

# Print to check the content of late_delivery_counts DataFrame
print(late_delivery_counts)
print('\n')

# Sort markets by the count of late deliveries in descending order
late_delivery_counts = late_delivery_counts.sort_values(by='Late Deliveries', ascending=False)

# Print to check the sorted DataFrame
print(late_delivery_counts)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(late_delivery_counts['Market'], late_delivery_counts['Late Deliveries'], color='goldenrod')
plt.ylabel('Count of Late Deliveries')
plt.title('Count of Late Deliveries by Market')
plt.xticks(rotation=360, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

print('We can see that the ratio of on-time deliveries to late deliveries. There are more late deliveries than on-time.')
```
